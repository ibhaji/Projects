**Importing Required Libraries**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

**Assign Columns Names**

```python
index_names = ['engine', 'cycle']
setting_names = ['setting_1', 'setting_2', 'setting_3']
sensor_names=[ "s1(Fan inlet temperature) (◦R)",
"s2(LPC outlet temperature) (◦R)",
"s3(HPC outlet temperature) (◦R)",
"s4(LPT outlet temperature) (◦R)",
"s5(Fan inlet Pressure) (psia)",
"s6(bypass-duct pressure) (psia)",
"s7(HPC outlet pressure) (psia)",
"s8(Physical fan speed) (rpm)",
"s9(Physical core speed) (rpm)",
"s10(Engine pressure ratio(P50/P2)",
"s11(HPC outlet Static pressure) (psia)",
"s12(Ratio of fuel flow to Ps30) (pps/psia)",
"s13(Corrected fan speed) (rpm)",
"s14(Corrected core speed) (rpm)",
"s15(Bypass Ratio) ",
"s16(Burner fuel-air ratio)",
"s17(Bleed Enthalpy)",
"s18(Required fan speed)",
"s19(Required fan conversion speed)",
"s20(High-pressure turbines Cool air flow)",
"s21(Low-pressure turbines Cool air flow)" ]
col_names = index_names + setting_names + sensor_names
```

**Creating Remaining Useful Life Column(RUL) to Predict the Engine Failure**

```python
dfr = df.groupby(['engine']).agg({'cycle':'max'})
dfr.rename(columns={'cycle':'life'},inplace=True)
df=df.merge(dfr,how='left',on=['engine'])
df['RUL']=df['life']-df['cycle']
df.drop(['life'],axis=1,inplace=True)
```

```python
df['RUL'][df['RUL']>200]=200
df
#The RUL prediction will be more accurate at the end of the engine's life, 
#therefore we are applying the above condition.
```

**Splitting the Data into Train and Test**

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

**Feature Scaling the Data using MinMaxScaler**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

**Fitting Various Regression Model to the splitted Data**

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X_train,y_train)
y_pred = model.predict(X_test)
print('R2-Score: '+ "{:.2%}".format(r2_score(y_test,y_pred)))
print('Root Mean Squared Error: '+ "{:.2f}".format(mean_squared_error(y_test,y_pred,squared=False)))
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor().fit(X_train,y_train)
y_pred = model.predict(X_test)
print('R2-Score: '+ "{:.2%}".format(r2_score(y_test,y_pred)))
print('Root Mean Squared Error: '+ "{:.2f}".format(mean_squared_error(y_test,y_pred,squared=False)))
```

```python
from sklearn.svm import SVR
model = SVR().fit(X_train,y_train)
y_pred = model.predict(X_test)
print('R2-Score: '+ "{:.2%}".format(r2_score(y_test,y_pred)))
print('Root Mean Squared Error: '+ "{:.2f}".format(mean_squared_error(y_test,y_pred,squared=False)))
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor().fit(X_train,y_train)
y_pred = model.predict(X_test)
print('R2-Score: '+ "{:.2%}".format(r2_score(y_test,y_pred)))
print('Root Mean Squared Error: '+ "{:.2f}".format(mean_squared_error(y_test,y_pred,squared=False)))
```

**By comparing above models, The Random forest regressor model has perfomed well with the score of 87%.**

**Implementing the Discretization Technique**

```python
df = pd.read_csv('train_FD001.txt',sep=' ',header=None,index_col=False,names=col_names)
```

```python
dfr = df.groupby(['engine']).agg({'cycle':'max'})
dfr.rename(columns={'cycle':'life'},inplace=True)
df=df.merge(dfr,how='left',on=['engine'])
```

```python
df['RUL']=df['life']-df['cycle']
df.drop(['life'],axis=1,inplace=True)
```

**Here we are using Equal-Frequency Discretization**

```python
def equal(x, nbin):
    nlen = len(x)
    return np.interp(np.linspace(0, nlen, nbin + 1),
                     np.arange(nlen),
                     np.sort(x))
n, bins, patches = plt.hist(list(df['RUL']), equal(list(df['RUL']), 3), edgecolor='black')
plt.show()
print(bins)
```

**As we can see the continuous data is tranformed into bins with each bin having the same number of records**

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

**Here we are splitting the data into three categories**

1.   Risk Zone
2.   Moderate Risk Zone
3.   Safe Zone





```python
y=[]
for i in df['RUL']:
    if i<=68:
        y.append(1)
    elif i>68 and i<=137:
        y.append(2)
    else:
        y.append(3)
```

```python
y=pd.DataFrame(y)
y.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

**Comparing Different Classification Models**

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression().fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print("Classification Report",(classification_report(y_test,y_pred)))
```

```python
from sklearn.svm import SVC
classifier = SVC().fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print("Classification Report",(classification_report(y_test,y_pred)))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier().fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print("Classification Report",(classification_report(y_test,y_pred)))
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier(n_neighbors=100).fit(X_train,y_train)
y_pred = classifier.predict(X_test)
print("Classification Report",(classification_report(y_test,y_pred)))
```

**By comparing the above models we conclude that the Random Forest Classifier has perfomed well with the accuracy of 77%**
